### Using label encoder to help in preprocessing the `RAINTODAY` an `RAINTOMMORROW`

Label Encoder: is used when we only have to featues to predict and they are not in a specific order

```python
label_encoder = LabelEncoder()

label_encoder_columns = ["RainToday", "RainTomorrow"]
for column in label_encoder_columns:
    data[column] = label_encoder.fit_transform(data[column])
```

### wont be using one hot encoder simply because it will tamper with features that has the same values; 
i will be using a function with a prefix in to help identify features values

```python
def column_prefix(data, column, prefix):
    return data[column].apply(lambda x: prefix + str(x))
```

```python
data["WindDir9am"] =  column_prefix(data, "WindDir9am", "9_")
data["WindDir3pm"] = column_prefix(data, "WindDir9am", "3_")
```

```python
data
```

```python
def onehot_encoder(data, columns):
    for column in columns:
        dummies = pd.get_dummies(data[column])
        data = pd.concat([data, dummies], axis=1)
        data.drop(column, axis=1,inplace=True)
    return data
        
```

```python
categorical_features = ["Location","WindGustDir", "WindDir9am","WindDir3pm"]
data=  onehot_encoder(data, categorical_features)
```

```python
def impute_means(data, columns):
    for column in columns:
        data[column]= data[column].fillna(data[column].mean())
```

```python
na_columns =['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
       'Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm']
impute_means(data, na_columns)
```
